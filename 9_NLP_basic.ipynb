```python
import nltk
nltk.download('stopwords')
```

### 1. NLTK에서 불용어 확인하기

```python
stop_words_list = stopwords.words('english')
print('불용어 개수:', len(stop_words_list))
print('불용어 10개 출력:', stop_words_list[:10])
```

### 2. NLTK를 통해 불용어 제거

```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []
for word in word_tokens:
    if word not in stop_words:
        result.append(word)

print('불용어 제거 전:', word_tokens)
print('불용어 제거 후:', result)
```

### 3. 한국어에서 불용어 제거하기

```python
okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전:', word_tokens)
print('불용어 제거 후:', result)
```

# 9-4 Regular Expression

## 1. 정규 표현식 문법과 모듈 함수
### 1) 정규 표현식 문법
파이썬에서는 Regular Expression 모듈 <span style='color:orange'>re</span>를 지원  

- 특수문자

|특수 문자|설명|
|:---|:---|
|.|한 개의 임의의 문자를 나타냄(\n제외)|
|?|앞의 문자가 존재할 수도 있고, 존재하지 않을 수도 있음(문자가 0개 or 1개)|
|*|앞의 문자가 무한개로 존재할 수도 있고, 존재하지 않을 수도 있음(문자가 0개 이상)|
|+|앞의 문자가 최소 한 개 이상 존재|
|^|뒤의 문자열로 문자열이 시작|
|$|앞의 문자열로 문자열이 끝남|
|{숫자}|숫자만큼 반복|
|{숫자1, 숫자2}|숫자1 이상, 숫자2 이하만큼 반복(?, *, +를 이걸로 대체 가능)|
|{숫자,}|숫자 이상만큼 반복|
|[]|대괄호 안의 문자들 중 한 개의 문자와 매치|
|[^문자]|해당 문자를 제외한 문자를 매치|
|l|AlB와 같이 쓰이며 A또는 B의 의미|

- 역슬래쉬(\)를 이용해 자주 쓰이는 문자 규칙들

|문자 규칙|설명|
|---|---|
|\\\\\ |역 슬래쉬 문자 자체를 의미|
|\\\d|모든 숫자를 의미([0-9]와 의미 동일)|
|\\\D|숫자를 제외한 모든 문자를 의미([^0-9])|
|\\\s|공백을 의미([\t\n\r\f\v])|
|\\\S|공백을 제외한 문자를 의미([^\t\n\r\f\v])|
|\\\w|문자 또는 숫자를 의미([a-zA-Z0-9])|
|\\\W|문자 또는 숫자가 아닌 문자를 의미([^a-zA-Z0-9])|

### 2) 정규표현식 모듈 함수

|모듈 함수|설명|
|---|---|
|re.compile()|정규표현식을 컴파일하는 함수. 즉, 파이썬에게 전해주는 역할|
|re.search()|문자열 전체에 대해 정규표현식과 매치되는지 검색|
|re.match()|문자열의 처음이 정규표현식과 매치되는지 검색|
|re.split()|정규 표현식을 기준으로 문자열을 분리하여 리스트로 리턴|
|re.findall()|문자열에서 정규 표현식과 매치되는 모든 경우의 문자열을 찾아 리스트로 리턴. 없으면 빈 리스트 리턴|
|re.finditer()|문자열에서 정규표현식과 매치되는 모든 경우의 문자열에 대한 이터레이터 객체를 리턴|
|re.sub()|문자열에서 정규표현식과 일치하는 부분에 대해 다른 문자열로 대체|

## 2. 단어 집합(Vocabulary) 생성
단어 집합(vocabulary): 중복을 제거한 텍스트의 총 단어 집합(set)을 의미  

'네이버 영화 리뷰 분류하기' 데이터를 이용한 실습  
총 20만개 영화 리뷰를 긍정(1), 부정(0)으로 레이블링한 데이터

```python
import urllib.request
import pandas as pd
from konlpy.tag import Mecab
from nltk import FreqDist
import numpy as np
import matplotlib.pyplot as plt
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt")

data = pd.read_table('ratings.txt')
data[:10]
```

```python
print('전체 샘플의 수: {}'.format(len(data)))
```

```python
sample_data = data[:100]        # 임의로 100개만 저장
```

```python
# 정규 표현식을 이용한 데이터 정제
# 한글과 공백을 제외하고 모두 제거
sample_data['document'] = sample_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "")
sample_data[:10]
```

```python
# 토큰화
# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
# mecab 을 이용한 형태소 분석
tokenizer = Mecab()
tokenized = []

for sentence in sample_data['document']:
    temp = tokenizer.morphs(sentence)
    temp = [word for word in temp if word not in stopwords]
    tokenized.append(temp)
```

```python
print(tokenized[:10])
```

```python
np.hstack(tokenized)        # 옆으로 쭉 이어 붙임
```

```python
# 단어 집합 생성
vocab = FreqDist(np.hstack(tokenized))          # 단어를 key로 단어에 대한 빈도수가 value
print('단어 집합의 크기: {}'.format(len(vocab)))
```

```python
vocab['재밌']
```

```python
# 등장 빈도수 사위 500개의 단어만 단어 집합으로 저장
vocab_size = 500

vocab = vocab.most_common(vocab_size)
print('단어 집합의 크기: {}'.format(len(vocab)))
```

## 3. 각 단어에 고유한 정수 부여
enumerate()는 순서 있는 자료형을 입력으로 받아 index를 순차적으로 함께 return  
0과 1은 다른 용도로 사용할 예정 => 2부터 501까지 순차적으로 index 부여

```python
word_to_index = {word[0] : index + 2 for index, word in enumerate(vocab)}

word_to_index['pad'] = 1
word_to_index['unk'] = 0
```

```python
# 기존 훈련 데이터에서 각 단어를 고유한 정수로 부여하는 작업 진행
encoded = []

for line in tokenized:
    temp = []
    for w in line:
        try:
            temp.append(word_to_index[w])   # 글자를 해당되는 정수로 변환
        except KeyError:    # 단어 집합에 없는 단어일 경우 unk로 대체
            temp.append(word_to_index['unk'])   # unk의 index로 변환

    encoded.append(temp)
```

## 4. 길이가 다른 문장들을 모두 동일한 길이로 바꿔주는 padding
길이가 다른 리뷰들을 모두 동일한 길이로 바꿔주는 작업  
'pad'를 이용

```python
max_len = max(len(l) for l in encoded)
print('리뷰 최대의 길이: {}'.format(max_len))
print('리뷰의 최소 길이: {}'.format(min(len(l) for l in encoded)))
print('리뷰의 평균 길이: {}'.format(sum(map(len, encoded))/len(encoded)))

plt.hist([len(s) for s in encoded], bins=50)
plt.xlabel('length of sample')
plt.ylabel('number of sample')
plt.show()
```

```python
# 가장 긴 리뷰 길이 62으로 통일
for line in encoded:
    if len(line) < max_len:
        line += [word_to_index['pad']] * (max_len-len(line))
```
